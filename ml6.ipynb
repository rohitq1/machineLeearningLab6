```python
import numpy as np
import matplotlib.pyplot as plt
def step_function(x):
    return 1 if x >= 0 else 0
# Define the AND gate inputs and corresponding outputs
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
W = np.array([10, 0.2, -0.75])
# Learning rate
alpha = 0.05
# Number of epochs
epochs = 100
errors = []
# Training the perceptron
for epoch in range(epochs):
    total_error = 0
    for i in range(len(X)):
        # Forward pass
        x_input = np.insert(X[i], 0, 1)  # Inserting bias
        z = np.dot(W, x_input)
        y_pred = step_function(z)
        # Calculate error   
        error = y[i] - y_pred
        total_error += error ** 2
        # Update weights
        W += alpha * error * x_input
    # Append total error to the error array
    errors.append(total_error)
plt.plot(range(epochs), errors)
plt.xlabel('Epochs')
plt.ylabel('Sum Square Error')
plt.title('Epochs vs Error')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Bi-Polar Step activation function
def bipolar_step_function(x):
    return 1 if x >= 0 else -1
# Sigmoid activation function
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))
# ReLU activation function
def relu_function(x):
    return max(0, x)
# AND gates (input and output)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
# Initial weights
W = np.array([10, 0.2, -0.75])
# Learning rate
alpha = 0.05
# Number of epochs
epochs = 100
# Error array to store the error after each epoch for each activation function
errors = {'Bi-Polar Step': [], 'Sigmoid': [], 'ReLU': []}
# Training the perceptron for each activation function
for activation_function in ['Bi-Polar Step', 'Sigmoid', 'ReLU']:
    for epoch in range(epochs):
        total_error = 0
        for i in range(len(X)):
            # Forward pass
            x_input = np.insert(X[i], 0, 1)  # Inserting bias
            z = np.dot(W, x_input)
            if activation_function == 'Bi-Polar Step':
                y_pred = bipolar_step_function(z)
            elif activation_function == 'Sigmoid':
                y_pred = sigmoid_function(z)
            elif activation_function == 'ReLU':
                y_pred = relu_function(z)
            # Calculate error
            error = y[i] - y_pred
            total_error += error ** 2
            # Update weights
            W += alpha * error * x_input
        errors[activation_function].append(total_error)
plt.figure(figsize=(10, 6))
for activation_function, error_values in errors.items():
    plt.plot(range(epochs), error_values, label=activation_function)
plt.xlabel('Epochs')
plt.ylabel('Sum Square Error')
plt.title('Epochs vs Error for Different Activation Functions')
plt.legend()
plt.grid(True)
plt.show()
```
